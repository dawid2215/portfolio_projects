## Load the data

### Subtask:
Load the data from the CSV file into a pandas DataFrame.


**Reasoning**:
The first step is to load the data from the CSV file into a pandas DataFrame.



```python
import pandas as pd

df = pd.read_csv('/content/solana.csv')
```

**Reasoning**:
Display the first few rows and the info of the DataFrame to verify the data loading and inspect its structure and data types.



```python
display(df.head())
display(df.info())
```

## Perform exploratory data analysis (eda)

### Subtask:
Analyze the distribution of numerical features using histograms.


## Perform exploratory data analysis (eda)

### Subtask:
Analyze the relationships between numerical features using a correlation heatmap.


**Reasoning**:
Calculate and visualize the correlation matrix for the numerical columns to understand their relationships.



```python
numerical_cols = ['Open', 'High', 'Low', 'Close', 'Volume', 'Market Cap']
plt.figure(figsize=(10, 8))
sns.heatmap(df[numerical_cols].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

## Data cleaning

### Subtask:
Handle missing values in the dataset.


**Reasoning**:
Check for missing values in each column of the DataFrame.



```python
display(df.isnull().sum())
```

## Data cleaning

### Subtask:
Handle outliers in the dataset.


**Reasoning**:
Identify outliers in the numerical columns using box plots as per the instructions.



```python
for col in numerical_cols:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=df[col])
    plt.title(f'Box Plot of {col}')
    plt.xlabel(col)
    plt.show()
```

## Data cleaning

### Subtask:
Check for duplicate data and handle it.


**Reasoning**:
Check for and handle duplicate rows in the DataFrame and then display the number of rows.



```python
duplicate_rows_count = df.duplicated().sum()
print(f"Number of duplicate rows before handling: {duplicate_rows_count}")

if duplicate_rows_count > 0:
    df.drop_duplicates(inplace=True)
    print("Duplicate rows removed.")
else:
    print("No duplicate rows found.")

print(f"Number of rows after handling duplicates: {len(df)}")
```

## Data visualization

### Subtask:
Create visualizations to explore the data and answer business questions. This may involve time series plots, scatter plots, or other relevant charts.


**Reasoning**:
Convert the 'Date' column to datetime objects and then create time series plots for 'Close' price and 'Volume', followed by a scatter plot of 'Close' vs. 'Volume'.



```python
df['Date'] = pd.to_datetime(df['Date'])

plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='Close', data=df)
plt.title('Close Price Over Time')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.show()

plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='Volume', data=df)
plt.title('Volume Over Time')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Volume', y='Close', data=df)
plt.title('Close Price vs. Volume')
plt.xlabel('Volume')
plt.ylabel('Close Price')
plt.show()
```
